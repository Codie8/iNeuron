```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns #Seaborn is used only to load dataset
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],columns= iris['feature_names'] + ['target1'])
df
```

```python
x = df["sepal length (cm)"]
y = df["sepal width (cm)"]
t = df["target1"]
t
```

```python
fig  = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x,y,c=t)
```

```python
x1 = df["petal length (cm)"]
y1 = df["petal width (cm)"]
t1 = df["target1"]
t1
```

```python
fig1  = plt.figure()
ax1 = fig1.add_subplot(111,projection='3d')
ax1.scatter(x1,y1,c=t1)
```
